# test.csv 만드는 코드

```python
import pandas as pd
from ultralytics import YOLO

MODEL_PATH = "/home/work/Project/level2-objectdetection-cv-02/baseline/ultralytics/recycling/experiment2/weights/best.pt"
TEST_PATH = "/home/work/Project/level2-objectdetection-cv-02/dataset/test"
# YOLO 모델 로드
model = YOLO(MODEL_PATH)  # 훈련된 모델 경로

# 테스트 이미지에 대한 예측 수행
results = model(TEST_PATH, save=False, save_txt=False)
```

```python
results
```

```python
for result in results:
    for box in result.boxes:
        print(box)
```

```python

# 결과를 저장할 리스트
formatted_predictions = []

# 각 이미지의 예측 결과를 처리
for result in results:
    image_id = f"test/{result.path.split('/')[-1]}"  # 이미지 파일 이름 추출
    predictions = []
    
    for box in result.boxes:
        class_id = int(box.cls)
        conf = float(box.conf)
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        
        pred_string = f"{class_id} {conf:.6f} {x1:.5f} {y1:.5f} {x2:.5f} {y2:.5f}"
        predictions.append(pred_string)
    
    # 모든 예측을 하나의 문자열로 결합
    all_preds = ' '.join(predictions)
    formatted_predictions.append((all_preds, image_id))

# DataFrame 생성
results_df = pd.DataFrame(formatted_predictions, columns=['PredictionString', 'image_id'])

# CSV 파일로 저장
results_df.to_csv('test.csv', index=False)

print(results_df.head())
```
